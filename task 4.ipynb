# TASK 4  DECISION TREE

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target']=iris.target
```

```python
df
```

```python
df['features_names']=df['target'].apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x=df.drop(columns=df[['target','features_names']])
y=df['features_names']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
dt=DecisionTreeClassifier()
```

```python
model=dt.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
fl=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
fn=['setosa','virginica','versicolor']
tree.plot_tree(model,feature_names=fl,class_names=fn,filled=True)
```
